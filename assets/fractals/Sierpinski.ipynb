```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = plt.gca()
ax.set_facecolor("black")
color="white"

tris=[[[0,0],[1/2,np.sqrt(3)/2],[1,0]]]

def draw_tris(tris):
    for t in tris:
        a=t[0]
        b=t[1]
        c=t[2]
        x=[a[0],b[0],c[0]]
        y=[a[1],b[1],c[1]]
        ax.fill(x,y,'white')
        ax.plot(([a[0]], [b[0]]), ([a[1]], [b[1]]),linestyle='-',color=color,linewidth=1)
        ax.plot(([a[0]], [c[0]]), ([a[1]], [c[1]]),linestyle='-',color=color,linewidth=1)
        ax.plot(([b[0]], [c[0]]), ([b[1]], [c[1]]),linestyle='-',color=color,linewidth=1)        

    
        
        
def get_points(pts):
        a=pts[0]
        b=pts[1]
        c=pts[2]
        length=c[0]-a[0]
        alph=[a[0]+length/2*np.cos(np.pi/3),a[1]+length/2*np.sin(np.pi/3)]
        beta=[a[0]+length/2,a[1]]
        gamma=[alph[0]+beta[0]-a[0],alph[1]+beta[1]-a[1]]

        
        return([alph,beta,gamma])

def generate_tris(tris):
    new_tris=[]
    for t in tris:
        a=t[0]
        b=t[1]
        c=t[2]
        pts=get_points([a,b,c])
        new_tris.append([a,pts[0],pts[1]])
        new_tris.append([pts[1],pts[2],c])        
        new_tris.append([pts[0],b,pts[2]])        
    
    return new_tris
    

def sierpinski(n):
    global tris
    for i in range(n):
        tris=generate_tris(tris)
    draw_tris(tris)
        

i=7 #order
sierpinski(i)

fig.set_size_inches(6,6)
#ax.axis('off')
ax.set_xticks([],[])
#ax.set_ylim(-2.7,1)
ax.set_yticks([],[])
ax.set_aspect("equal")

fig.savefig("sierpinski"+str(i)+".pdf",bbox_inches='tight',dpi=500)
```
