```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np
```

```python
fig = plt.figure(figsize=(10,8))
ax = plt.gca()
color="black"



def draw_line(pos,length,slope,wdth):
    x1=pos[0]
    y1=pos[1]
    
    ax.plot(([x1], [x1+length*np.cos(slope)]), ([y1], [y1+length*np.sin(slope)]),linestyle='-',color=color,linewidth=wdth)

    return [x1+length*np.cos(slope),y1+length*np.sin(slope)]

def branch(pos,length,slope,wdth):
#    angle=np.random.rand()*np.pi*2
    angle=np.pi/8
    global lim
    
    length=length*0.66
    wdth=wdth*0.7
    
    if (length>lim):
        slopel=slope+angle
        posl=draw_line(pos,length,slopel,wdth)
        branch(posl,length,slopel,wdth)
        
       # slopem=slope
       # posm=draw_line(pos,length,slopem,wdth)
       # branch(posm,length,slopem,wdth)
        
        sloper=slope-angle
        posr=draw_line(pos,length,sloper,wdth)
        branch(posr,length,sloper,wdth)
        
        
pos=draw_line([0,0],1,np.pi/2,5)

lim=0.25/64
branch(pos,1,np.pi/2,5)

    

ax.set_ylim([0.0,3])
ax.set_xlim([-5,5])
ax.set_xticks([],[])
ax.set_yticks([],[])

fig.savefig("tree"+str(lim)+".pdf",bbox_inches='tight',dpi=500)
```
