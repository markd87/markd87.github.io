```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

```python
def nn_graph(nodes,colors=[[1.0, 0.5, 0.5],[0.5, 0.5, 1],[0.5, 0.7509803921568627, 0.5]],fname=None):
    """
    draws neural network graph with the number of nodes is each layer given by the "nodes" list,
    and the color of each layer's nodes given by a list of colors (optional).
    fname: file name to save the graph image.
    """
    
    nodes=[n for n in nodes if isinstance(n,int)]
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

    layers=len(nodes)
    mx=np.max(nodes)

    node_r=0.3

    plt.figure(figsize=(1.5*layers,mx))
    for i in range(layers):
        x=np.zeros(nodes[i])+2*i
        y=range(nodes[i])
        if (nodes[i]<mx):
            y+=(mx-nodes[i])/2
            
        #draw nodes
        for px,py in zip(x,y):
            circ = plt.Circle((px, py), radius=node_r, color=colors[i%len(colors)])
            ax.add_patch(circ)
        
        #draw connecting lines
        if i>0:
            for a in prevy:
                for b in y:
                    ax.plot([2*(i-1)+node_r,2*i-node_r],[a,b],'gray')
                    #ax.margins(0.2, 0.2)
        
        prevy=y
        
    
    ax.axis('off')
    ax.set_aspect("equal")
    if fname!=None:
        fig.savefig(fname,bbox_inches='tight',dpi=300)
        print('figure saved as: ',fname) 
```

```python
nodes=[3,4,2]
nn_graph(nodes,colors,'nn.png')
```
